```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("transactions.csv")
df
```

```python
#What is the total sales amount across all transactions?

total_sales = df["t_amt"].sum()
print(total_sales)
```

```python
df["t_date"] = pd.to_datetime(df["t_date"])

df["month"] = df["t_date"].dt.month
df["quarter"] = df["t_date"].dt.quarter
```

```python
#Which month had the highest total transaction amount?
month_sales = df.groupby("month")["t_amt"].sum()
print(month_sales.idxmax())
```

```python
#What is the average transaction amount per customer?
avg_customer = df.groupby("cust_id")["t_amt"].mean()
print(avg_customer)
```

```python
#What is the trend of total sales over months?

trend = df.groupby("month")["t_amt"].sum()
print(trend)
```

```python
#What is the highest single transaction amount recorded?

print(df["t_amt"].max())
```

```python
#What is the highest single transaction amount recorded?

service_sales = df.groupby("services")["t_amt"].sum()
print(service_sales.idxmax()) #idxmax() tells you where the biggest value is
```

```python
#Which service category contributes the most to total revenue?

product_sales = df.groupby("products_used")["t_amt"].sum()
print(product_sales.idxmax())
```

```python
#Which product generated the highest revenue?

print(df.groupby("services")["t_amt"].mean())
```

```python
#How does the average transaction amount vary between services?
print(df.groupby("services")["t_amt"].mean())
```

```python
#How many unique customers made purchases?
print(df["cust_id"].nunique())
```

```python
#Which customers have spent the most overall?
customer_total = df.groupby("cust_id")["t_amt"].sum()
print(customer_total.sort_values(ascending=False))
```

```python
#What is the average number of transactions per customer?
transactions = df.groupby("cust_id").size()
print(transactions.mean())
```

```python
#Are there customers who made multiple purchases in different categories?
multi = df.groupby("cust_id")["services"].nunique()
print((multi > 1).sum())
```

```python
#What percentage of customers are repeat buyers?
counts = df["cust_id"].value_counts()
repeat = counts > 1
percent = repeat.mean() * 100
print(percent)
```

```python
#Which product category has the highest total sales?
print(df.groupby("services")["t_amt"].sum().idxmax())
```

```python
#Which services are most popular by transaction count?
print(df["services"].value_counts())
```

```python
#For each service, which product type is purchased most frequently?
group = df.groupby(["services", "products_used"]).size()
print(group)
```

```python
#What is the average transaction amount per product type?
print(df.groupby("products_used")["t_amt"].mean())
```

```python
#Are there specific services where customers spend significantly more?
print(df.groupby("services")["t_amt"].mean().sort_values(ascending=False))
```

```python
#Which state has the highest total sales?
state_sales = df.groupby("state")["t_amt"].sum()
print(state_sales.idxmax())
```

```python
#Which city records the highest number of transactions?
print(df["city"].value_counts().idxmax())
```

```python
#What is the average spending per transaction in each state?
print(df.groupby("state")["t_amt"].mean())
```

```python
#Are certain services more popular in specific states?
print(df.groupby(["state", "services"]).size())
```

```python
#Which states buy the most Outdoor Recreation products?
outdoor = df[df["services"] == "Outdoor Recreation"]
print(outdoor.groupby("state")["t_amt"].sum())
```

```python
#Compare average spending between California and Texas customers.
subset = df[df["state"].isin(["California", "Texas"])]
print(subset.groupby("state")["t_amt"].mean())
```

```python
#Which quarter of the year saw the highest sales?
quarter_sales = df.groupby("quarter")["t_amt"].sum()
print(quarter_sales.idxmax())
```

```python
#How does total sales vary month by month?
print(df.groupby("month")["t_amt"].sum())
```

```python
#What is the total number of transactions per month?
print(df.groupby("month")["t_amt"].sum())
```

```python
#Is there a particular season when sports equipment sales spike?
print(df.groupby("month").size())
```

```python
#How many transactions were done using credit?
print(df.groupby("month")["t_amt"].sum().idxmax())
```

```python
#What is the total revenue generated from credit transactions?
print((df["t_details"] == "credit").sum())
```

```python
#Is there a difference in average spending between credit and debit customers?
credit = df[df["t_details"] == "credit"]
print(credit["t_amt"].sum())
```
